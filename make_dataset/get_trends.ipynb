```
pdc = prod_sess.groupby(['CLAC3_NM','PD_BRA_NM']).agg({'CLNT_ID':'count','PD_BUY_CT':'sum', 'SEARCH_CNT':'sum','TOT_PAG_VIEW_CT':'mean','TOT_SESS_HR_V':'mean'})
```

```
pdc['BUY/SEARCH'] = pdc['PD_BUY_CT'] / pdc['SEARCH_CNT']
```

```
def normalize1(x):
  return (x-x.mean())/(x.std())
def normalize2(x):
  return (x-x.min())/(x.max()-x.min())
```

```
pdc['TOT_PAG_VIEW'] = normalize1(pdc['TOT_PAG_VIEW_CT'])
pdc['TOT_SESS_HR'] = normalize1(pdc['TOT_SESS_HR_V'])
pdc['PD_WAITING'] = normalize2((pdc['TOT_PAG_VIEW']+pdc['TOT_SESS_HR'])/2)
pdc['PD_BUY'] = normalize1(pdc['PD_BUY_CT'])
pdc['SEARCH'] = normalize1(pdc['SEARCH_CNT'])
pdc['CLNT'] = normalize1(pdc['CLNT_ID'])
```

```
pdc.describe()
```

```
pdcdf = pdc[['CLNT','PD_BUY','SEARCH','PD_WAITING']]
```

```
pdcdf.head()
```

```
pdcdf['W_HIGHLOW'] = np.where(pdcdf['PD_WAITING'] >= pdcdf.PD_WAITING.mean(),'high','low')
pdcdf['B_HIGHLOW'] = np.where(pdcdf['PD_BUY'] >= pdcdf['PD_BUY'].mean(),'high','low')
pdcdf['S_HIGHLOW'] = np.where(pdcdf['SEARCH'] >= pdcdf['SEARCH'].mean(),'high','low')
pdcdf['C_HIGHLOW'] = np.where(pdcdf['CLNT'] >= pdcdf['CLNT'].mean(),'high','low')
```

```
pdcdf['TYPE'] = np.where(pdcdf['W_HIGHLOW'] == 'high',
                         (np.where(pdcdf['B_HIGHLOW'] == 'high',
                                   np.where(pdcdf['S_HIGHLOW'] == 'high','hhh','hhl'),
                                    np.where(pdcdf['S_HIGHLOW'] == 'high','hlh','hll'))),
                         (np.where(pdcdf['B_HIGHLOW'] == 'high',
                                   np.where(pdcdf['S_HIGHLOW'] == 'high','lhh','lhl'),
                                    np.where(pdcdf['S_HIGHLOW'] == 'high','llh','lll'))))
```

```
pdcdf['TYPE'] = np.where(pdcdf['C_HIGHLOW'] == 'high',
                         (np.where(pdcdf['B_HIGHLOW'] == 'high','hh','hl')),
                         (np.where(pdcdf['B_HIGHLOW'] == 'high','lh','ll')))
```
